To Perform Testing of Stationary

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# H0: it is not stationary
# H1: it is stationary

def adf_test(sales):
  result = adfuller(sales)
  print('ADF Statistic: %f' % result[0])
  print('p-value: %f' % result[1])
  print(f'# Lags used: {result[2]}')
  print(f'No of observations used: {result[3]}')
  if result[1] <= 0.05:
    print('we reject the null hypothesis, The series is stationary')
  else:
    print('Not enough statistical evidence to reject null hypothesis, The series is not stationary')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

def analyze_stationarity(series: pd.Series):
    # ADF Test
    print("### Augmented Dickey-Fuller Test ###")
    result = adfuller(series.dropna())  # Handle NaN values if any
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    print(f"# Lags used: {result[2]}")
    print(f"No of observations used: {result[3]}")
    
    if result[1] <= 0.05:
        print("Conclusion: We reject the null hypothesis (H0). The series is stationary.")
    else:
        print("Conclusion: Not enough statistical evidence to reject H0. The series is not stationary.")

    # Rolling Statistics Visualization
    rolling_mean = series.rolling(window=12).mean()
    rolling_std = series.rolling(window=12).std()

    plt.figure(figsize=(14, 8))
    plt.plot(series, label="Original Data", color="blue")
    plt.plot(rolling_mean, label="Rolling Mean (12 periods)", color="red", linestyle="--")
    plt.plot(rolling_std, label="Rolling Std Dev (12 periods)", color="green", linestyle="--")
    plt.title("Rolling Mean and Standard Deviation")
    plt.xlabel("Time")
    plt.ylabel("Values")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Decomposition
    print("\n### Seasonal Decomposition ###")
    decomposition = seasonal_decompose(series, model="additive", period=12)
    decomposition.plot()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Prepare the data
def prepare_data(pre_train, pre_test):
    # Assuming 'y' is the target column in your pre_train DataFrame
    train_data = pre_train['y'].values.reshape(-1, 1)
    test_data = pre_test['y'].values.reshape(-1, 1)
    
    # Normalize data
    scaler = MinMaxScaler()
    train_scaled = scaler.fit_transform(train_data)
    test_scaled = scaler.transform(test_data)
    
    return train_scaled, test_scaled, scaler

# Create sequences for LSTM input
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])  # Predicting the next value after n_steps
    return np.array(X), np.array(y)

# Build and train the LSTM model
def build_lstm_model(n_steps, X_train, y_train):
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(n_steps, 1)),
        Dense(1)  # Output layer
    ])
    model.compile(optimizer='adam', loss='mse')
    
    model.fit(X_train, y_train, epochs=20, verbose=1)
    return model

# Make predictions using the trained model
def predict_lstm(model, X_test, scaler, n_steps, test_scaled):
    predicted = model.predict(X_test)
    predicted = scaler.inverse_transform(predicted)  # Inverse scaling to get original values

    # Extract actual values (inverse scaled)
    actual_values = scaler.inverse_transform(test_scaled[n_steps:])
    
    return predicted, actual_values

# Evaluate the model performance
def evaluate_model(actual_values, predicted):
    mae = mean_absolute_error(actual_values, predicted)
    mse = mean_squared_error(actual_values, predicted)
    rmse = np.sqrt(mse)

    print(f'LSTM MAE: {mae}')
    print(f'LSTM MSE: {mse}')
    print(f'LSTM RMSE: {rmse}')

# Plot the forecast
def plot_forecast(actual_values, predicted):
    # Convert results to DataFrame for visualization
    forecast_df = pd.DataFrame({
        'Actual': actual_values.flatten(),
        'Predicted': predicted.flatten()
    }, index=pd.date_range(start='2023-01-01', periods=len(actual_values), freq='D'))  # Adjust the start date accordingly
    
    forecast_df.plot(figsize=(12, 6))
    plt.xlabel('Date')
    plt.ylabel('Quantity Sold')
    plt.title('Actual vs. Predicted Quantity Sold for "The Barbecue Chicken Pizza"')
    plt.legend()
    plt.show()

# Forecast for a future period (7 days)
def forecast_future(model, last_sequence, scaler, n_steps):
    # Use the last sequence to predict the next values (e.g., 7 days forecast)
    future_predictions = []
    current_input = last_sequence.reshape(1, n_steps, 1)
    
    for _ in range(7):  # Forecast for the next 7 days
        pred = model.predict(current_input)
        future_predictions.append(pred[0, 0])
        current_input = np.append(current_input[:, 1:, :], pred.reshape(1, 1, 1), axis=1)

    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
    
    return future_predictions

# Main function to execute the entire workflow
def run_lstm_forecast(pre_train, pre_test, n_steps=3):
    # Prepare data
    train_scaled, test_scaled, scaler = prepare_data(pre_train, pre_test)
    
    # Create sequences for LSTM input
    X_train, y_train = create_sequences(train_scaled, n_steps)
    X_test, y_test = create_sequences(test_scaled, n_steps)
    
    # Reshaping for LSTM: (samples, time steps, features)
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
    
    # Build and train the model
    model = build_lstm_model(n_steps, X_train, y_train)
    
    # Make predictions
    predicted, actual_values = predict_lstm(model, X_test, scaler, n_steps, test_scaled)
    
    # Evaluate the model
    evaluate_model(actual_values, predicted)
    
    # Plot forecast results
    plot_forecast(actual_values, predicted)
    
    # Forecast future values (next 7 days)
    last_sequence = test_scaled[-n_steps:]
    future_predictions = forecast_future(model, last_sequence, scaler, n_steps)
    
    # Print the future forecast
    print("Future forecast for the next 7 days:", future_predictions)

# Run the LSTM forecast
run_lstm_forecast(pre_train, pre_test)
```
